```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("/content/drive/MyDrive/ML Datasets/insurance.csv")
df = pd.DataFrame(a)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df["children"].value_counts()
```

```python
df
```

```python
df["sex"].value_counts()
```

```python
df["sex"] = df["sex"].replace("male", 1)
df["sex"] = df["sex"].replace("female", 0)
```

```python
df["smoker"].value_counts()
```

```python
df["smoker"] = df["smoker"].replace("no", 1)
df["smoker"] = df["smoker"].replace("yes", 0)
```

```python
df["region"].value_counts()
```

```python
df["region"] = df["region"].replace("southeast", 0)
df["region"] = df["region"].replace("southwest", 1)
df["region"] = df["region"].replace("northwest", 2)
df["region"] = df["region"].replace("northeast", 3)
```

```python
df
```

```python
X = df.drop(columns="charges").values
```

```python
Y = df["charges"].values
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.4, random_state=2, shuffle=True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train, Y_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
from sklearn import metrics
```

```python
X_train_rsq = metrics.r2_score(X_train_prediction, Y_train)
print(X_train_rsq)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
X_test_rsq = metrics.r2_score(X_test_prediction, Y_test)
print(X_test_rsq)
```

```python
import gradio as gr
import numpy as np
import joblib  # to load model and scaler

# Load the trained model and scaler
model = joblib.load("insurance_model.pkl")     # replace with your filename
scaler = joblib.load("insurace_scalar.pk1")   # replace with your scaler file

# Prediction function
def predict_insurance(age, sex, bmi, children, smoker, region):
    try:
        # Convert smoker and region to numeric labels
        sex = 1 if sex == "Male" else 0
        smoker = 1 if smoker == "no" else 0
        region_map = {"Southeast": 0, "Southwest": 1, "Northwest": 2, "Northeast": 3}
        region = region_map[region]

        # Prepare input data
        input_data = np.array([age, sex, bmi, children, smoker, region]).reshape(1, -1)
        input_scaled = scaler.transform(input_data)

        # Predict insurance charges
        prediction = model.predict(input_scaled)[0]
        return f"💵 Predicted Insurance Cost: ${prediction}"

    except Exception as e:
        return f"❌ Error: {str(e)}"

# Gradio UI
gr.Interface(
    fn=predict_insurance,
    inputs=[
        gr.Number(label="Age"),
        gr.Radio(choices=["Male", "Female"], label="Sex"),
        gr.Number(label="BMI"),
        gr.Number(label="Number of Children"),
        gr.Radio(choices=["Yes", "No"], label="Smoker"),
        gr.Dropdown(choices=["Southeast", "Southwest", "Northwest", "Northeast"], label="Region")
    ],
    outputs="text",
    title="🧾 Insurance Cost Prediction",
    description="Enter the details below to estimate your insurance charges using a trained Random Forest model."
).launch()
```
